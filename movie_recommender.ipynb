```python
## genre is String of lists
## String of lists: lets convert it to Lists

import ast

ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')

## see how it changed the thing
```

```python
def convert(text):
    arr = []
    for i in ast.literal_eval(text):
        arr.append(i['name'])

    return arr
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies.iloc[0]['cast']

## has a huge dataset
## lets just take the top 3 actors and thats it
```

```python
def convert_cast(text):
    arr= []
    count = 0
    for i in ast.literal_eval(text):
        if count<3:
            arr.append(i['name'])
            count+=1
    return arr
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.head(1)
```

```python
movies['crew'][0]

## we will just use the director tag from this
```

```python
def convert_crew(text):
    arr = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            arr.append(i['name'])
    return arr
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
movies.head(1)
```

```python
movies['overview'][0]

## its in the string format, lets convert it to list
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies['overview'][0]
```

```python
movies.head(1)

## names have gaps between them, so will be treated as 2 diff entities
## so lets replace space with undeescore
```

```python
arr = ['cast','crew','genres','keywords']
for j in arr:
    movies[j] = movies[j].apply(lambda x: [i.replace(" ","_") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags','genres']]
new_df.head(5)
```

```python
## lets make the tags to string

new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head(3)
```

```python
new_df['tags'][0]
```

```python
## its always recommended to put your text in lower case.
## so lets  convert to lower case

new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
## we need to stem the whole tags thing
## since dance same as dancing but treated as 2 diff entities

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]

## marine changed to marin
## unique to uniqu
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
vector.shape

## 4806 movies
## 5000 words
```

```python
cv.get_feature_names_out()[:20]
```

```python
len(cv.get_feature_names_out())
```

```python
## since we are dealing in the higher dimesnions
## euclidia distance will fail as it wont be that reliable in higher dimensions
### simply we called this as curse of dimensionality

## so for that we will be using the cosine distance
## that is angle between the two vectors


## less the distance: more the similarity means less angle between two vectors


from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vector)

## here we are taking distace(cosine angle) of each movie with otheer movie
```

```python
cosine_similarity(vector).shape

## since 4806 movies: so 4806 distanced between one and rest of the movies
```

```python
similarity = cosine_similarity(vector)

## this matrix will have diagonal as 1
## since similarity will be 1 of each movie with itself
```

```python

## index of any movie
index = new_df[new_df['title']=='Batman Begins'].index[0]
print(index)

sorted(similarity[index],reverse=True)
## but index is not being shown

sorted(list(enumerate(similarity[0])),reverse=True, key = lambda x:x[1])
## index values also shown
## and also sorted on the basis of the second values and not on the index values
```

```python
## recommendation function
## find the similarity vector in the similarity object for the particular movie
## and the 5 least distanced will be the movies similar to it

def recommend(movie):
    index = new_df[new_df['title']==movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True, key = lambda x:x[1])
    for i in distance[1:11]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend("The Dark Knight Rises")
```

```python
from sklearn.metrics import ndcg_score

# Evaluation of Recommendation System using NDCG and MRR Metrics

# Create dummy true and predicted labels for demonstration
y_true = np.random.randint(0, 2, 100)  # Binary relevance labels (1 = relevant, 0 = not relevant)
y_pred = np.random.random(100)  # Predicted relevance scores

# Custom MRR (Mean Reciprocal Rank) function
def mean_reciprocal_rank(y_true, y_pred):
    # Calculate the Mean Reciprocal Rank (MRR) for predictions
    ranks = [1 / (np.where(np.argsort(y_pred)[::-1] == i)[0][0] + 1) for i in np.where(y_true == 1)[0]]
    return np.mean(ranks) if ranks else 0

# Calculate NDCG Score using sklearn
ndcg_value = ndcg_score([y_true], [y_pred])

# Calculate MRR using the custom function
mrr_value = mean_reciprocal_rank(y_true, y_pred)

# Print Evaluation Metrics
print("Evaluation Metrics:")
print(f"NDCG Score: {ndcg_value:.4f}")
print(f"MRR: {mrr_value:.4f}")
```

SAVE THE MODEL

```python
import pickle
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Load the dataset
movies_path = "tmdb_5000_movies.csv"
movies_df = pd.read_csv(movies_path)

# 2. Extract the movie titles
movies = movies_df['title'].dropna().tolist()  # Drop missing titles

# 3. Create a TF-IDF similarity matrix
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies)
similarity_matrix = cosine_similarity(tfidf_matrix)

# 4. Save the similarity matrix using pickle
model_path = "movie_recommender_model.pkl"
with open(model_path, 'wb') as f:
    pickle.dump(similarity_matrix, f)

print(f"Model saved at {model_path}")
```

USE THE SAVED MODEL

```python
import pickle
import pandas as pd
import numpy as np

# 1. Load the similarity matrix from the saved .pkl file
model_path = "movie_recommender_model.pkl"
with open(model_path, 'rb') as f:
    similarity_matrix = pickle.load(f)

# Confirm the type of the loaded model
print(f"Loaded model type: {type(similarity_matrix)}")

# 2. Load the movies dataset
movies_path = "tmdb_5000_movies.csv"
movies_df = pd.read_csv(movies_path)
movies_df = movies_df[['id', 'title']].dropna()  # Ensure no missing titles

# 3. Function to get recommendations by movie name
def get_recommendations_by_name(movie_name, similarity_matrix, top_n=10):
    """
    Recommend top N movies based on the given movie name.
    """
    # Find the index of the movie that matches the name
    movie_idx = movies_df[movies_df['title'].str.lower() == movie_name.lower()].index
    if len(movie_idx) == 0:
        print(f"Movie '{movie_name}' not found!")
        return []

    # Get the similarity scores for the selected movie
    movie_idx = movie_idx[0]
    similarity_scores = similarity_matrix[movie_idx]
    top_indices = np.argsort(similarity_scores)[::-1][1:top_n + 1]  # Exclude the original movie

    # Retrieve the recommended movie titles
    return movies_df.iloc[top_indices][['title']]

# 4. Example usage
movie_name = "The Dark Knight"  # Change this to test other movies
recommended_movies = get_recommendations_by_name(movie_name, similarity_matrix, top_n=5)

# 5. Display the recommendations
if not recommended_movies.empty:
    print(f"Top 5 movies recommended for '{movie_name}':")
    print(recommended_movies.to_string(index=False))
```
